```python
#By Dibyajyoti Dhar
def show_dice(x):
  if x<=42.5:
    print('\u2685', end ="")
  elif x<=42.5*2:
    print('\u2684', end ="")
  elif x<=42.5*3:
    print('\u2683', end ="")
  elif x<=42.5*4:
    print('\u2682', end ="")
  elif x<=42.5*5:
    print('\u2681', end ="")
  elif x<=42.5*6:
    print('\u2680', end ="")

def getshade(x):
  if x<=85:
    print('\u2593', end ="")
  elif x<=85*2:
    print('\u2592', end ="")
  elif x<=85*3:
    print('\u2591', end ="")

def newgetshade(x):
  if x<=51:
    print('\u2588', end ="")
  elif x<=51*2:
    print('\u2593', end ="")
  elif x<=51*3:
    print('\u2592', end ="")
  elif x<=51*4:
    print('\u2591', end ="")
  elif x<=51*5:
    print(' ', end ="")

from google.colab.patches import cv2_imshow
import cv2
import numpy as np
img = cv2.imread('pp.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2. equalizeHist(gray)
alpha = 1.5 # Contrast control (1.0-3.0)
beta = -10 # Brightness control (0-100)

gray = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)
dim = 100
resized_gray = cv2.resize(gray, (dim,dim), interpolation=cv2.INTER_AREA) 
cv2_imshow(gray)

z = np.zeros([dim//2,dim], dtype=int)

for i in range(0, len(z)):
  for j in range(0, len(z[0])):
    z[i][j]=((int)(resized_gray[i*2][j])+(int)(resized_gray[i*2+1][j]))/2


for i in range(0, len(z)):
  for j in range(0, len(z[0])):
    newgetshade(z[i][j])
  print('')


```
